```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')   # gas price($/gallon) in different countiries in different year

gas.head(5)
```

### Line Graph

```python
plt.style.use('default')

# Giving title for given figure
plt.title('Gas Prices Over Time (in USD)', fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 18}, loc = 'center')

# Creating line chart for each country
country_names = list(gas.columns[1:])
for i in country_names:
    plt.plot(gas['Year'], gas[i], '.-', label = i)

# Displaying year values in x-axis
plt.xticks(gas['Year'][::3])

# Show x- axis as Year and y-axis as gas prices($/gallon)
plt.xlabel('Year',  fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 13}, loc = 'right')
plt.ylabel('US Dollars',  fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 13}, loc = 'center')

# Showing which line refer to which country and adding legend outside figure
plt.legend(title = 'Countries:', bbox_to_anchor = (1,1))

# adding grid lines to given figure 
plt.grid()

# To display given line graph
plt.show()
```

### Load FIFA Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa = fifa.drop(columns = ['Unnamed: 0'])
```

```python
fifa.head()
```

### Histogram

```python
# Creating bins with given list of values
b = [40,50,60,70,80,90,100]

# Creating Histogram
plt.hist(fifa['Overall'], bins = b, color = '#abcdef', edgecolor = 'k', linewidth = 2)

# Displaying x values for skill level
plt.xticks(b)

# Labelling Title, x-axis and Y-axis
plt.title('Distribution Of Player Skills in FIFA',  fontdict = {'family' : 'sans serif', 'weight' : 'normal', 'size' : 18}, loc = 'left')
plt.xlabel('Skill Level-->', fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 13}, loc = 'right')
plt.ylabel('Number Of Players-->', fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 13}, loc = 'bottom')

# Convert Y-axis scale to linear scale
plt.yscale('linear')

# Display a Histogram
plt.show()
```

### Pie Charts

```python
# Find no. of Football Players with Left Footer Preference
left = fifa.loc[fifa['Preferred Foot'] == 'Left']['Preferred Foot'].count()

# Find no. of Football Players with Right Footer Preference
right = fifa.loc[fifa['Preferred Foot'] == 'Right']['Preferred Foot'].count()

# Adding labels and specify color to each label
lst_labels = ['Left', 'Right']
lst_colors = ['#abcdef', '#aaabcc']

# Creating Pie Chart
pc = plt.pie([left, right], labels = lst_labels, colors = lst_colors, autopct = '%0.2f%%', radius = 1.2, explode = [0.05, 0.05], textprops = {'size' : 18})

# Set color to specific wedge
pc[0][0].set(edgecolor = 'k', linewidth = 2)
pc[0][1].set(edgecolor = 'k', linewidth = 2)

# Labelling Title for given figure
plt.title('Foot Preferences Of FIFA Players:', fontdict = {'family' : 'sans serif', 'weight' : 'normal', 'size' : 18}, loc = 'center')

# To Display given pie chart
plt.show()
```

```python
# Extract weight(in lbs) of each Football Player
fifa['Weight_Numeric'] = fifa['Weight'].str[0:3].astype(float)

# Finding no. of Football Players whose weights are less than 125 lbs
light = fifa.loc[fifa['Weight_Numeric'] < 125]['Weight_Numeric'].count()

# Finding no. of Football Players whose weights are in 125 - 150 lbs
light_medium = fifa.loc[(fifa['Weight_Numeric'] >= 125) & (fifa['Weight_Numeric'] < 150)]['Weight_Numeric'].count()

# Finding no. of Football Players whose weights are in 150 - 175 lbs
medium = fifa.loc[(fifa['Weight_Numeric'] >= 150) & (fifa['Weight_Numeric'] < 175)]['Weight_Numeric'].count()

# Finding no. of Football Players whose weights are in 175 - 200 lbs
medium_heavy = fifa.loc[(fifa['Weight_Numeric'] >= 175) & (fifa['Weight_Numeric'] < 200)]['Weight_Numeric'].count()

# Finding no. of Football Players whose weights are more than 200 lbs
heavy = fifa.loc[fifa['Weight_Numeric'] >= 200]['Weight_Numeric'].count()

weights = [light, light_medium, medium, medium_heavy, heavy]
lst_labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
lst_explodes = [0.4, 0.2, 0.05, 0, 0.4]

# Display title for given figure
plt.title('Weight Distributiuon Of FIFA Players (in lbs)', fontdict = {'family' : 'sans serif', 'weight' : 'normal', 'size' : 18}, loc = 'center')

# Creating Pie Chart
pc = plt.pie(weights, labels = lst_labels, autopct = '%.2f%%', pctdistance = 0.7, labeldistance = 1.1, explode  = lst_explodes)

# To set color to specific wedge 
pc[0][2].set(edgecolor = 'k', linewidth = 2)
pc[0][3].set(edgecolor = 'k', linewidth = 2)

# To group and summarise weight in pictorial way
plt.legend(title = 'Weight Division:' , bbox_to_anchor = (1.7,1))

# To display a pie chart
plt.show()
```

### Box and Whiskers Chart

```python
plt.figure(figsize = (8,6))

# Finding Overall values of all football players to specific club
Barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
Madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
Revs = fifa.loc[fifa['Club'] == 'New England Revolution']['Overall']
Fortuna = fifa.loc[fifa['Club'] == 'Fortuna Dsseldorf']['Overall']

lst_labels = ['FC_Barcelona', 'Real_Madrid', 'NIE_evolution', 'Fortuna_Dsseldorf']

# Create Boxplot
bp = plt.boxplot([Barcelona, Madrid, Revs, Fortuna], labels = lst_labels, patch_artist = True, medianprops = {'linewidth' : 2})

# Set color to each PathPatch object
for box in bp['boxes']:
    box.set(edgecolor = '#4286f4', linewidth = 2, facecolor = '#e0e0e0')

# headline given to figure
plt.title('Professional Soccer Team Comparison', fontdict = {'family' : 'sans serif', 'weight' : 'normal', 'size' : 18}, loc = 'center')

# label to y-axis
plt.ylabel('FIFA Overall Rating-->', fontdict = {'family' : 'serif', 'weight' : 'bold', 'size' : 13}, loc = 'bottom')

# To display a boxplot
plt.show()
```
